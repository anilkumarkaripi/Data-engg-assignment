#####  3.Write a Python program to print all the prime numbers between 1 and N (inclusive). Where N is a positive integer and an input from user.

```python
def is_prime(x: int) -> bool:
    if x < 2:
        return False
    if x in (2, 3):
        return True
    if x % 2 == 0 or x % 3 == 0:
        return False
    
    i = 5
    while i * i <= x:
        if x % i == 0 or x % (i + 2) == 0:
            return False
        i += 6
    return True


def main():
    try:
        N = int(input("Enter a positive integer N: ").strip())
        if N <= 0:
            print("N must be a positive integer.")
            return
    except ValueError:
        print("Invalid input. Please enter a positive integer.")
        return

    primes = [num for num in range(2, N+1) if is_prime(num)]

    if primes:
        print("Prime numbers between 1 and", N, "are:")
        print(*primes)
    else:
        print("There are no prime numbers between 1 and", N)


if __name__ == "__main__":
    main()
```

##### 4 .Write a function merge_sorted_lists(list1, list2) that takes two sorted lists of integers, list1 and list2 and merges them into a single sorted list efficiently. One way is to simple concatenate the lists and sort. But think of a better approach.
#### Example:
##### list1 = [1, 4, 5]
##### list2 = [1, 3, 4]
##### merged = merge_sorted_lists(list1, list2)
##### print(merged) #####

```python
def merge_sorted_lists(list1, list2):
	    i, j = 0, 0
	    merged = []
	  
	    while i < len(list1) and j < len(list2):
	        if list1[i] <= list2[j]:
	            merged.append(list1[i])      
	            i += 1
	        else:
	            merged.append(list2[j])
	            j += 1
	
	    # Append any leftovers
	    if i < len(list1):
	        merged.extend(list1[i:])
	    if j < len(list2):
	        merged.extend(list2[j:])
	
	    return merged
	
	
list1 = [1, 4, 5]
list2 = [1, 3, 4]
merged = merge_sorted_lists(list1, list2)
print(merged)
```

```python
def is_anagram_sorted(str1: str, str2: str) -> bool:
	    return sorted(str1.lower()) == sorted(str2.lower())
	
	
print(is_anagram_sorted("listen", "silent"))  # True
print(is_anagram_sorted("hello", "world"))    # False
```

##### 7.Write a function rotate_list(lst, k) that rotates a given list lst to the right by k positions in-place. Assume k is a non-negative integer and can be greater than the length of the list.

```python

def rotate_list(lst, k):
    n = len(lst)
    if n == 0:
        return
    k %= n  
    lst.reverse()  
    lst[:k] = reversed(lst[:k])  
    lst[k:] = reversed(lst[k:])  

my_list = [1, 2, 3, 4, 5]
rotate_list(my_list, 2)
print(my_list)
    	   
```

#####  8.you have a list of dictionaries, where each dictionary represents a product with keys "name", "price", and "category". Write a function group_by_category(products) that takes this list and returns a dictionary where the keys are the unique categories and the values are lists of product names belonging to that category


```python
from collections import defaultdict

def group_by_category(products):
    grouped = defaultdict(list)
    
    for item in products:
        name = item.get("name")
        category = item.get("category")
        
        if name is not None and category is not None:
            grouped[category].append(name)
    
    return dict(grouped)

products = [
    {"name": "Laptop",  "price": 1200, "category": "Electronics"},
    {"name": "Book",    "price": 25,   "category": "Books"},
    {"name": "Tablet",  "price": 300,  "category": "Electronics"},
    {"name": "Fiction", "price": 15,   "category": "Books"},
]

grouped_products = group_by_category(products)
print(grouped_products)
```

##### 9.Write a Python program that prints the first n rows of Pascal's triangle. n is a positive integer and an user input. To know about Pascal's triangle see the wikipedia article:


```python
def pascal_triangle(n: int):
    if n <= 0:
        print("n must be a positive integer.")
        return

    row = [1]
    for i in range(n):
        print(" ".join(map(str, row)))

        row = [0] + row + [0]
        row = [row[j] + row[j + 1] for j in range(len(row) - 1)]


def main():
    try:
        n = int(input("Enter a positive integer n: ").strip())
    except ValueError:
        print("Invalid input. Please enter a positive integer.")
        return

    pascal_triangle(n)


if __name__ == "__main__":
    main()
```

##### 10.Write a function reverse_the_words(s) which will take a string s as input and reverses the words in the string.


```python
def reverse_the_words(s):
    return ' '.join(s.split()[::-1])


s = "Python is awesome and challenging"
r = reverse_the_words(s)
print(r)  # Output: challenging and awesome is Python
```
